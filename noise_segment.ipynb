```python
import cv2
from skimage.util import random_noise
import matplotlib.pyplot as plt
from PIL import ImageChops
```

```python
import numpy as np
pic=plt.imread('images.jpg')
ax=plt.subplots(figsize=(20,10))
noisy_im=random_noise(pic)
plt.subplot(121)
plt.imshow(pic)
plt.subplot(122)
plt.imshow(noisy_im)
```

```python
from skimage.restoration import denoise_tv_chambolle
pic1=plt.imread('noisy1.jpg')
ax=plt.subplots(figsize=(20,10))
denoise_im=denoise_tv_chambolle(pic1, multichannel=True)
plt.subplot(121)
plt.imshow(pic1)
plt.subplot(122)
plt.imshow(denoise_im)
```

```python
from skimage.restoration import denoise_bilateral
pic1=plt.imread('noisy1.jpg')
ax=plt.subplots(figsize=(20,10))
denoise_im=denoise_bilateral(pic1, multichannel=True)
plt.subplot(121)
plt.imshow(pic1)
plt.subplot(122)
plt.imshow(denoise_im)
```

```python
from skimage.segmentation import slic
from skimage.color import label2rgb
pic=plt.imread('images.jpg')
ax=plt.subplots(figsize=(20,10))
segments=slic(pic, n_segments=6000)
seg_im=label2rgb(segments, pic, kind='avg')
plt.subplot(121)
plt.imshow(pic)
plt.subplot(122)
plt.imshow(seg_im)
```
